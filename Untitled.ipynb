```python
import tensorflow as tf
import numpy as np

# Heat equation parameters
alpha = 0.01  # Thermal diffusivity

# Define the domain and conditions
x_min, x_max = 0.0, 1.0  # Spatial domain
t_min, t_max = 0.0, 1.0  # Time domain

# Initial condition: u(x, 0) = sin(pi * x)
def initial_condition(x):
    return np.sin(np.pi * x)

# Boundary conditions: u(0, t) = 0, u(1, t) = 0
def boundary_condition_left(t):
    return 0.0

def boundary_condition_right(t):
    return 0.0

# Generate training data
n_x = 50
n_t = 50
x = np.linspace(x_min, x_max, n_x)
t = np.linspace(t_min, t_max, n_t)
x, t = np.meshgrid(x, t)
x_train = x.flatten()[:, None]
t_train = t.flatten()[:, None]

# Convert to TensorFlow tensors
x_train_tf = tf.convert_to_tensor(x_train, dtype=tf.float32)
t_train_tf = tf.convert_to_tensor(t_train, dtype=tf.float32)

# PINN model
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(2,)),
        tf.keras.layers.Dense(10, activation="tanh"),
        tf.keras.layers.Dense(10, activation="tanh"),
        tf.keras.layers.Dense(1, activation=None)
    ])
    return model

model = create_model()

# Physics-informed loss function
def compute_loss(model, x, t):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch([x, t])
        u = model(tf.concat([x, t], axis=1))  # u(x, t)
        u_t = tape.gradient(u, t)  # du/dt
        u_x = tape.gradient(u, x)  # du/dx
        u_xx = tape.gradient(u_x, x)  # d^2u/dx^2
    
    # Residual loss (PDE loss)
    residual = u_t - alpha * u_xx
    mse_pde = tf.reduce_mean(tf.square(residual))

    # Initial condition loss
    x_ic = tf.convert_to_tensor(np.linspace(x_min, x_max, n_x)[:, None], dtype=tf.float32)
    t_ic = tf.convert_to_tensor(np.full((n_x, 1), t_min), dtype=tf.float32)
    u_ic_pred = model(tf.concat([x_ic, t_ic], axis=1))
    u_ic_true = tf.convert_to_tensor(initial_condition(x_ic.numpy()), dtype=tf.float32)
    mse_ic = tf.reduce_mean(tf.square(u_ic_pred - u_ic_true))

    # Boundary condition loss
    t_bc = tf.convert_to_tensor(np.linspace(t_min, t_max, n_t)[:, None], dtype=tf.float32)
    x_bc_left = tf.convert_to_tensor(np.full((n_t, 1), x_min), dtype=tf.float32)
    x_bc_right = tf.convert_to_tensor(np.full((n_t, 1), x_max), dtype=tf.float32)
    u_bc_left = model(tf.concat([x_bc_left, t_bc], axis=1))
    u_bc_right = model(tf.concat([x_bc_right, t_bc], axis=1))
    mse_bc = tf.reduce_mean(tf.square(u_bc_left - boundary_condition_left(t_bc))) + \
             tf.reduce_mean(tf.square(u_bc_right - boundary_condition_right(t_bc)))

    return mse_pde + mse_ic + mse_bc

# Training loop
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
epochs = 1000

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x_train_tf, t_train_tf)
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))

    if epoch % 200 == 0:
        print(f"Epoch {epoch}: Loss = {loss.numpy()}")

# Testing and visualization
import matplotlib.pyplot as plt

# Generate test data
x_test = np.linspace(x_min, x_max, n_x)
t_test = np.linspace(t_min, t_max, n_t)
x_test, t_test = np.meshgrid(x_test, t_test)
x_test_flat = x_test.flatten()[:, None]
t_test_flat = t_test.flatten()[:, None]
x_test_tf = tf.convert_to_tensor(x_test_flat, dtype=tf.float32)
t_test_tf = tf.convert_to_tensor(t_test_flat, dtype=tf.float32)

# Predict u(x, t)
u_pred = model(tf.concat([x_test_tf, t_test_tf], axis=1)).numpy()
u_pred = u_pred.reshape(n_x, n_t)

# Plot the results
plt.figure(figsize=(8, 6))
plt.contourf(x_test, t_test, u_pred, levels=100, cmap="hot")
plt.colorbar(label="Velocity")
plt.xlabel("x")
plt.ylabel("t")
plt.title("Predicted Solution u(x, t)")
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define the physical constants and problem domain
nu = 0.01  # Kinematic viscosity
x_min, x_max = 0.0, 1.0  # Spatial domain in x
y_min, y_max = 0.0, 1.0  # Spatial domain in y
t_min, t_max = 0.0, 1.0  # Time domain

# Number of training points in x, y, and t
n_x, n_y, n_t = 50, 50, 50

# Initial condition for velocity (u and v) and pressure
def initial_condition(x, y):
    u_init = np.sin(np.pi * x) * np.cos(np.pi * y)
    v_init = -np.cos(np.pi * x) * np.sin(np.pi * y)
    p_init = np.zeros_like(x)  # Pressure initial condition (zero for simplicity)
    return u_init, v_init, p_init

# Boundary conditions for velocity at the boundaries (no-slip)
def boundary_conditions(x, y):
    return 0.0, 0.0  # u = 0, v = 0 at the boundaries

# Generate training data
x = np.linspace(x_min, x_max, n_x)
y = np.linspace(y_min, y_max, n_y)
t = np.linspace(t_min, t_max, n_t)

x_train, y_train, t_train = np.meshgrid(x, y, t)
x_train = x_train.flatten()[:, None]
y_train = y_train.flatten()[:, None]
t_train = t_train.flatten()[:, None]

# Convert to TensorFlow tensors
x_train_tf = tf.convert_to_tensor(x_train, dtype=tf.float32)
y_train_tf = tf.convert_to_tensor(y_train, dtype=tf.float32)
t_train_tf = tf.convert_to_tensor(t_train, dtype=tf.float32)

# Define the PINN model
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(3,)),  # Input: (x, y, t)
        tf.keras.layers.Dense(10, activation="tanh"),
        tf.keras.layers.Dense(10, activation="tanh"),
        tf.keras.layers.Dense(3, activation=None)  # Output: (u, v, p)
    ])
    return model

model = create_model()

# Physics-Informed Loss Function
def compute_loss(model, x, y, t):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch([x, y, t])
        uvp = model(tf.concat([x, y, t], axis=1))  # u(x, y, t), v(x, y, t), p(x, y, t)
        u, v, p = uvp[:, 0:1], uvp[:, 1:2], uvp[:, 2:3]  # Separate u, v, p components
        
        # Compute gradients for Navier-Stokes equations
        u_t = tape.gradient(u, t)  # du/dt
        v_t = tape.gradient(v, t)  # dv/dt
        u_x = tape.gradient(u, x)  # du/dx
        v_x = tape.gradient(v, x)  # dv/dx
        u_y = tape.gradient(u, y)  # du/dy
        v_y = tape.gradient(v, y)  # dv/dy
        
        # Second derivatives for diffusion term
        u_xx = tape.gradient(u_x, x)  # d^2u/dx^2
        v_xx = tape.gradient(v_x, x)  # d^2v/dx^2
        u_yy = tape.gradient(u_y, y)  # d^2u/dy^2
        v_yy = tape.gradient(v_y, y)  # d^2v/dy^2
        
        # Continuity equation for incompressible flow: div(u) = 0
        p_x = tape.gradient(p, x)  # dp/dx
        p_y = tape.gradient(p, y)  # dp/dy

    # Residuals for Navier-Stokes equations (incompressible flow)
    u_res = u_t + u * u_x + v * u_y + (-1) * (u_xx + u_yy) / nu
    v_res = v_t + u * v_x + v * v_y + (-1) * (v_xx + v_yy) / nu
    
    # Continuity equation: div(u) = 0
    continuity_res = p_x + p_y

    # Boundary condition loss (no-slip, u = 0, v = 0)
    u_bc, v_bc = boundary_conditions(x, y)
    bc_loss = tf.reduce_mean(tf.square(u - u_bc)) + tf.reduce_mean(tf.square(v - v_bc))
    
    # Initial condition loss
    u_ic, v_ic, p_ic = initial_condition(x, y)
    ic_loss = tf.reduce_mean(tf.square(u - u_ic)) + tf.reduce_mean(tf.square(v - v_ic)) + tf.reduce_mean(tf.square(p - p_ic))
    
    # Total loss: PDE loss + boundary conditions + initial condition + continuity
    pde_loss = tf.reduce_mean(tf.square(u_res)) + tf.reduce_mean(tf.square(v_res)) + tf.reduce_mean(tf.square(continuity_res))
    
    return pde_loss + bc_loss + ic_loss

# Optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Training loop
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x_train_tf, y_train_tf, t_train_tf)
    
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss.numpy()}")

# Test the trained model
x_test = np.linspace(x_min, x_max, n_x)
y_test = np.linspace(y_min, y_max, n_y)
t_test = np.linspace(t_min, t_max, n_t)

x_test, y_test, t_test = np.meshgrid(x_test, y_test, t_test)
x_test = x_test.flatten()[:, None]
y_test = y_test.flatten()[:, None]
t_test = t_test.flatten()[:, None]

x_test_tf = tf.convert_to_tensor(x_test, dtype=tf.float32)
y_test_tf = tf.convert_to_tensor(y_test, dtype=tf.float32)
t_test_tf = tf.convert_to_tensor(t_test, dtype=tf.float32)

# Predict u, v, and p
uvp_pred = model(tf.concat([x_test_tf, y_test_tf, t_test_tf], axis=1)).numpy()
u_pred = uvp_pred[:, 0].reshape(n_x, n_y, n_t)
v_pred = uvp_pred[:, 1].reshape(n_x, n_y, n_t)
p_pred = uvp_pred[:, 2].reshape(n_x, n_y, n_t)
```
